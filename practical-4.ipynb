```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print("array a: ",a)
print("array b: ",b)

c = np.add(a, b)
print("Addition: ",c)

c = np.subtract(a, b)
print("Subtraction(a,b): ",c)

c = np.subtract(b, a)
print("Subtraction(b,a): ",c)

c = np.multiply(a, b)
print("Multiplication: ",c)

c = np.divide(a, b)
print("Division(a,b): ",c)

c = np.add(b, a)
print("Division(b,a): ",c)
```
