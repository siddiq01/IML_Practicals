**Write a Pandas program to create a line plot of the opening, closing stock
prices of given company between two specific dates.**

```python
# Siddiq Badlani
# 216270307034

import pandas as pd
from matplotlib import pyplot as plt

file = pd.read_csv('/kaggle/input/nifty50-stock-market-data/RELIANCE.csv')

start = "2005-07-28"
end = "2005-08-05"

data = file[(file['Date'] >= start) & (file['Date'] <= end)]
opening = data['Open']
closing = data['Close']
plt.plot(data['Date'], opening, color='green',label="Open")
plt.plot(data['Date'], closing,color='red', label="Close")
plt.title('Opening and Closing Stock Prices')
plt.legend()
plt.grid()
plt.show()
```
